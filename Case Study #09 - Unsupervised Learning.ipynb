# Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scaled_data = scalar.fit_transform(data)
scaled_data = pd.DataFrame(scaled_data, columns = data.columns)
scaled_data.describe()
```

# PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca.fit(scaled_data)
x_pca = pca.transform(scaled_data)
x_pca.shape
```

```python
pca.components_
```

```python
plt.figure(figsize =(8, 6))
plt.scatter(x_pca[:, 0], x_pca[:, 1])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

# K Means Clustering

```python
from sklearn.cluster import KMeans
ks=range(1,6)
inertia = []
for k in ks:
    model = KMeans(n_clusters = k)
    model.fit(scaled_data)
    inertia.append(model.inertia_)
```

```python
plt.plot(ks,inertia)
plt.xlabel('No of Clusters')
plt.ylabel('Inertia')
plt.show()
```

number of clusters can be taken as 3

```python
model_kmeans = KMeans(n_clusters=3)
model_kmeans.fit(scaled_data)
labels_sc = model_kmeans.predict(scaled_data)
print(labels_sc)
```

```python
centroids = model_kmeans.cluster_centers_
centroids_pca = pca.transform(centroids)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(x_pca[:, 0], x_pca[:, 1], c=labels_sc, cmap="brg")
plt.scatter(x=centroids_pca[:,0], y=centroids_pca[:,1], marker="x", s=100, linewidths=2, color="black")

plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```

```python
data['KMean_ClusterID']= pd.Series(labels_sc)
data.head()
```
