# Visualizations using the dataset

Sepal_width vs sepal_length

```python
plt.figure(figsize=(6,4))
plt.scatter(data['SL'],data['SW'],s=10)
plt.title('plot of Sepal_length vs Sepal_width')
plt.xlabel('SL')
plt.ylabel('SW')
```

```python
plt.figure(figsize=(10,8))
plt.subplot(2,1,1)
data1 = data[data['Classification'] == 'Iris-setosa']
plt.scatter(data1['PL'],data1['PW'],label='Iris-setosa',s=14,c='red',marker='*')
data1 = data[data['Classification'] == 'Iris-virginica']
plt.scatter(data1['PL'],data1['PW'],label='Iris-virginica',s=14,c='blue',marker='*')
data1 = data[data['Classification'] == 'Iris-versicolor']
plt.scatter(data1['PL'],data1['PW'],label='Iris-versicolor',s=14,c='green',marker='*')
plt.title('plot of Petal length vs petal width',fontsize=10)
plt.xlabel('PL',fontsize=10)
plt.ylabel('PW',fontsize=10)

plt.subplot(2,1,2)
data1 = data[data['Classification'] == 'Iris-setosa']
plt.scatter(data1['SL'],data1['SW'],s=14,label='Iris-setosa',c='red',marker='*')
data1 = data[data['Classification'] == 'Iris-virginica']
plt.scatter(data1['SL'],data1['SW'],s=14,label='Iris-virginica',c='blue',marker='*')
data1 = data[data['Classification'] == 'Iris-versicolor']
plt.scatter(data1['SL'],data1['SW'],s=14,label='Iris-versicolor',c='green',marker='*')
plt.title('plot of Sepal length vs Sepal width',fontsize=10)
plt.xlabel('SL',fontsize=10)
plt.ylabel('SW',fontsize=10)

plt.tight_layout()
```
